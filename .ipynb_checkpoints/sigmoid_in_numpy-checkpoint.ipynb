# The Sigmid Function 

<h1>1) What is the sigmoid function?</h1>

<p>
  The <strong>sigmoid function</strong> is a math function that takes any number and
  <strong>squeezes it</strong> into a value between:
</p>

<ul>
  <li><strong>0 and 1</strong></li>
</ul>

<p>
  That’s why it’s very common in <strong>Machine Learning</strong>, especially for
  <strong>binary classification</strong> (Yes/No, 0/1).
</p>

<hr />

<h1>2) The sigmoid formula</h1>

<p><strong>Sigmoid formula:</strong></p>

<ul>
  <li><code>sigmoid(x) = 1 / (1 + e^(-x))</code></li>
</ul>

<p><strong>Where:</strong></p>

<ul>
  <li><code>x</code> is your input number</li>
  <li><code>e</code> is a special constant (approximately 2.718)</li>
  <li><code>e^(-x)</code> means "e power negative x"</li>
</ul>

<hr />

<h1>3) What happens to different inputs?</h1>

<h2>If x is very large positive (example: x = 10)</h2>

<ul>
  <li><code>e^(-10)</code> becomes very small (almost 0)</li>
  <li>so the sigmoid becomes:</li>
</ul>

<ul>
  <li><code>sigmoid(10) ≈ 1 / (1 + 0) = 1</code></li>
</ul>

<p>So the output is close to <strong>1</strong>.</p>

<h2>If x is 0</h2>

<ul>
  <li><code>sigmoid(0) = 1 / (1 + e^(0))</code></li>
  <li><code>sigmoid(0) = 1 / (1 + 1)</code></li>
  <li><code>sigmoid(0) = 0.5</code></li>
</ul>

<p>So the output is <strong>0.5</strong>.</p>

<h2>If x is very negative (example: x = -10)</h2>

<ul>
  <li><code>e^(-(-10)) = e^(10)</code> which is huge</li>
  <li>so:</li>
</ul>

<ul>
  <li><code>sigmoid(-10) ≈ 1 / (1 + huge) ≈ 0</code></li>
</ul>

<p>So the output is close to <strong>0</strong>.</p>

<h1>5) Why do we use sigmoid?</h1>

<p>Because it behaves like a <strong>probability</strong>:</p>

<ul>
  <li>output near <strong>0</strong> means "probably class 0"</li>
  <li>output near <strong>1</strong> means "probably class 1"</li>
  <li>output near <strong>0.5</strong> means "not sure"</li>
</ul>

<p><strong>Examples:</strong></p>

<ul>
  <li><code>sigmoid(2) ≈ 0.88</code> means about <strong>88% chance of class 1</strong></li>
  <li><code>sigmoid(-2) ≈ 0.12</code> means about <strong>12% chance of class 1</strong></li>
</ul>

<h1>6) Python code (slow explanation)</h1>

<h2>Sigmoid in NumPy</h2>

<pre><code>import numpy as np

def sigmoid(x):
    x = np.asarray(x, dtype=float)   # convert input to numpy array
    return 1 / (1 + np.exp(-x))      # sigmoid formula
</code></pre>

<h2>Example usage</h2>

<pre><code>print(sigmoid(0))            # 0.5
print(sigmoid(2))            # ~0.8808

</code></pre>

<h1>7) Important idea (shape of sigmoid curve)</h1>

<p>Sigmoid is an <strong>S-shaped curve</strong>:</p>

<ul>
  <li>negative numbers go close to 0</li>
  <li>0 becomes exactly 0.5</li>
  <li>positive numbers go close to 1</li>
</ul>


!

The sigmoid function is one of the simplest and most widely used activation functions in machine learning. It takes any real number and maps it smoothly into the range (0, 1).

**Properties:**

- Shape: It has an “S-shaped” curve. Very negative inputs push the output close to 0, while very positive inputs push it close to 1.
- Centering: At input 𝑥 = 0, the output is exactly 0.5.
- Smoothness: The curve is continuous and differentiable everywhere, which makes it useful for optimization.
- Interpretation: Because the output is always between 0 and 1, it’s often treated as a probability.
  
Used in logistic regression, binary classification output layers, and historically in hidden layers.

```python
import numpy as np

def sigmoid(x):
    # vectorized sigmoid function
    x = np.asarray(x, dtype=float)
    return 1/(1+np.exp(-x))

print(sigmoid(-2))
print(sigmoid(3))
print(sigmoid(0))            # 0.5
print(sigmoid(2))            # ~0.8808
print(sigmoid([-2, 0, 2]))   # array outputs
```

<h1>1) What does <code>np.exp(-x)</code> mean?</h1>

<h2>A) What is <code>exp</code>?</h2>

<p><code>np.exp(a)</code> means:</p>

<ul>
  <li><strong>e</strong> raised to the power <strong>a</strong></li>
  <li>mathematically: <code>e^a</code></li>
</ul>

<p><strong>Example:</strong></p>

<ul>
  <li><code>np.exp(1) = e^1 ≈ 2.718</code></li>
  <li><code>np.exp(2) = e^2 ≈ 7.389</code></li>
</ul>

<h2>B) So what is <code>np.exp(-x)</code>?</h2>

<p>It means:</p>

<ul>
  <li><code>e^(-x)</code></li>
</ul>

<p><strong>Example if <code>x = 3</code>:</strong></p>

<ul>
  <li><code>np.exp(-3) = e^-3 ≈ 0.0498</code> (a small number)</li>
</ul>

<p><strong>Example if <code>x = -3</code>:</strong></p>

<ul>
  <li><code>np.exp(-(-3)) = np.exp(3) = e^3 ≈ 20.085</code> (a big number)</li>
</ul>

<p><strong>So:</strong></p>

<ul>
  <li>when <code>x</code> is positive, <code>exp(-x)</code> becomes small</li>
  <li>when <code>x</code> is negative, <code>exp(-x)</code> becomes large</li>
</ul>

<h1>2) Plot sigmoid using matplotlib</h1>

<p>Here is a simple plot:</p>


```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    x = np.asarray(x, dtype=float)
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 200)  # numbers from -10 to 10
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show()
```

<h2>What you will see</h2>

<ul>
  <li>Left side (negative <code>x</code>) → near <strong>0</strong></li>
  <li>Middle (<code>x = 0</code>) → <strong>0.5</strong></li>
  <li>Right side (positive <code>x</code>) → near <strong>1</strong></li>
</ul>

<h1>3) Sigmoid in Logistic Regression (simple explanation)</h1>

<p>
  Logistic regression is used when the answer is <strong>two choices</strong>, like:
</p>

<ul>
  <li>Pass / Fail</li>
  <li>Spam / Not spam</li>
  <li>Sick / Not sick</li>
</ul>

<h2>Step 1: Model makes a score (any number)</h2>

<p>
  Logistic regression first computes a <strong>linear score</strong>:
</p>

<ul>
  <li><code>z = w*x + b</code></li>
</ul>

<p><strong>Where:</strong></p>

<ul>
  <li><code>w</code> = weight</li>
  <li><code>x</code> = input feature</li>
  <li><code>b</code> = bias</li>
</ul>

<p>
  This <code>z</code> can be any number: negative, positive, big, small.
</p>

<h2>Step 2: Turn the score into a probability</h2>

<p>Now we apply sigmoid:</p>

<ul>
  <li><code>p = sigmoid(z)</code></li>
</ul>

<p>
  Now <code>p</code> is between <strong>0</strong> and <strong>1</strong>.
</p>

<p><strong>So we can say:</strong></p>

<ul>
  <li><code>p = 0.9</code> → 90% chance of class 1</li>
  <li><code>p = 0.1</code> → 10% chance of class 1</li>
</ul>

<h2>Step 3: Make the final decision</h2>

<p>
  Usually we use <strong>0.5</strong> as the decision boundary:
</p>

<ul>
  <li>if <code>p &gt;= 0.5</code> → predict <strong>1</strong></li>
  <li>if <code>p &lt; 0.5</code> → predict <strong>0</strong></li>
</ul>

<h2>Small example (very easy)</h2>

<p>Let’s say:</p>

<ul>
  <li><code>z = 2</code></li>
</ul>

<p>Then:</p>

<ul>
  <li><code>sigmoid(2) ≈ 0.88</code></li>
</ul>

<p>So the model says:</p>

<ul>
  <li>✅ “Class 1 with 88% confidence”</li>
</ul>

<p>If:</p>

<ul>
  <li><code>z = -2</code></li>
</ul>

<p>Then:</p>

<ul>
  <li><code>sigmoid(-2) ≈ 0.12</code></li>
</ul>

<p>So the model says:</p>

<ul>
  <li>✅ “Class 0 is more likely”</li>
</ul>


<h1>1) Sigmoid output as probability</h1>

<p>
  The sigmoid function turns any number into a value between <strong>0 and 1</strong>.
</p>

<p>
  This value can be interpreted as the <strong>probability</strong> that the answer belongs to class 1.
</p>

<p>
  So if <code>sigmoid(z) = 0.88</code>, it means:<br>
  “There’s an 88% chance this input belongs to class 1.”
</p>

<hr />

<h1>2) Confidence</h1>

<p><strong>Confidence</strong> = how sure the model is about its prediction.</p>

<ul>
  <li>Higher sigmoid output → more confident that it’s class 1.</li>
  <li>Lower sigmoid output → more confident that it’s class 0.</li>
</ul>

<p><strong>Examples:</strong></p>

<ul>
  <li><code>sigmoid(2) ≈ 0.88</code> → high confidence for class 1</li>
  <li><code>sigmoid(-2) ≈ 0.12</code> → high confidence for class 0</li>
</ul>

<hr />

<h1>3) More likely</h1>

<p>“More likely” just compares the probability to 0.5:</p>

<ul>
  <li>If <code>sigmoid(z) &gt; 0.5</code> → class 1 is more likely than class 0</li>
  <li>If <code>sigmoid(z) &lt; 0.5</code> → class 0 is more likely than class 1</li>
</ul>

<p><strong>Examples:</strong></p>

<ul>
  <li><code>sigmoid(2) ≈ 0.88</code> → &gt; 0.5 → class 1 is more likely</li>
  <li><code>sigmoid(-2) ≈ 0.12</code> → &lt; 0.5 → class 0 is more likely</li>
</ul>

<hr />

<h2>✅ Quick way to remember</h2>

<ul>
  <li>Sigmoid value near 1 → confident in class 1</li>
  <li>Sigmoid value near 0 → confident in class 0</li>
  <li>Sigmoid value ≈ 0.5 → uncertain, “both classes equally likely”</li>
</ul>
